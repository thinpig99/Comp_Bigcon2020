## 내부데이터

```python
def load_data(filenm):
    df = pd.read_excel(filenm, header = None)
    columns = df.loc[1].tolist()
    df.columns = columns
    df = df[2:]
    return df
```

### [특성처리_0]  전처리

- [1] 특정 컬럼 결측치 처리 : '노출(분)', '취급액'
- [2] 날짜 변수 분해 : 날짜-> 년/월/일/분/초
- [3] 변수 명 영어이름으로 바꾸기
- [4] 변수 dtype 조정

```python
## 특성 전처리

#[1] 결측치 처리 - 노출(분), 취급액
def missing_treat1(df):  #train 용
    df = df.dropna(subset = ['취급액'])
    df = df.fillna(method = 'pad')
    return df

def missing_treat2(df): #test용
    df = df.fillna(method = 'pad')
    df = df.fillna(0)
    return df


#[2] 날짜 컬럼 분해 (방송일시 -> 년/월/일/분/초)
def date2col(df):
    #날짜 값 처리
    df['방송일시'] = df['방송일시'].astype('datetime64[ns]')
    df['date'] = df['방송일시'].apply(lambda x: str(x)[:10])
    df['dt_year'] = df['방송일시'].apply(lambda x: int(str(x)[:4]))
    df['dt_month'] = df['방송일시'].apply(lambda x: int(str(x)[5:7]))
    df['dt_day'] = df['방송일시'].apply(lambda x: int(str(x)[8:10]))    
    df['dt_mmdd'] = df['방송일시'].apply(lambda x: str(x)[5:10].replace("-",""))
    df['dt_hour'] = df['방송일시'].apply(lambda x: int(str(x)[11:13]))
    df['dt_weekday1'] = df['방송일시'].apply(lambda x: datetime.weekday(x)) #0123456-월화수목금토일
    df['dt_weekday2'] = df['방송일시'].apply(lambda x: datetime.weekday(x)) #0123456-월화수목금토일/One-Hot인코딩용
    df['dt_minute'] = df['방송일시'].apply(lambda x: int(str(x)[14:16]))
    
    #경과 시간 컬럼 생성
    #df['dt_seconds'] = df['방송일시'].apply(lambda x: time.mktime(datetime.strptime(str(x)[11:19],'%H:%M:%S').timetuple()))
    old_time = datetime(2019,1,1,0,0,0)
    df['dt_elapsed time'] =df['방송일시'].apply(lambda x: (x-old_time).total_seconds()) #2019-01-01 이후
    df['dt_seconds'] = df['dt_hour']*60+df['dt_hour'] #자정 이후
     
    #요일값 One-Hot
    #df = pd.get_dummies(df,columns=['dt_weekday2']) #요일값 One-Hot 인코딩
    df = df.drop(['방송일시'], axis=1)
    
    return df

#[3] 컬럼명 한글 -> 영어 변경 
def colname2eng(df):
    df = df.rename({'노출(분)':'playtime_minute', '마더코드':'mother_code', 
                        '상품코드':'prd_code', '상품명':'prd_nm', '상품군':'prd_type',
                        '판매단가':'price', '취급액': 'total_price' } , axis = 'columns')
    return df

#[4] dtype 조정 : 노출(분), 판매단가, 취급액 : object -> int
def dtype_treat(df):
    df['playtime_minute'] = df['playtime_minute'].apply(lambda x: int(x))
    df['price'] = df['price'].apply(lambda x: int(x))
    df['total_price'] = df['total_price'].apply(lambda x: int(x))
    df['mother_code'] = df['mother_code'].apply(lambda x: str(x))
    df['prd_code'] = df['prd_code'].apply(lambda x: str(x))
    return df
```

```python
######   main  ###########
filenm_performance = 'data_2019.xlsx'
filenm_submission = 'submission_202006.xlsx'

#파일 데이터프레임으로 받아오기
performance = load_data(filenm_performance)
submission = load_data(filenm_submission)

#데이터프레임 전처리 
performance = dtype_treat(colname2eng(date2col(missing_treat1(performance))))
submission = dtype_treat(colname2eng(date2col(missing_treat2(submission))))
```

### [특성 처리_1] 범주형 변수 
- 상품군 : Label 인코딩
- 상품명 

 1) 일단 버리기!
 
 2) 상품군이 '무형'인 데이터 제거하기(결측치 처리에서 이미 제거됨)
 

```python
#상품명 = '무형'인 row 제거 -> 이미 '취급액'NaN 처리에서 제거(dropna )
#performance = performance[performance['prd_nm']!='무형']

#상품명 버리기
performance = performance.drop(['prd_nm'], axis=1) 

#상품군 Label Encoding
encoder_prdnm = LabelEncoder()
encoder_prdnm.fit(performance['prd_type'].values)
performance['prd_type_encoding'] = encoder_prdnm.transform(performance['prd_type'])
performance = performance.drop(['prd_type'], axis=1) 

performance.head()
```

### [특성 처리_2] 특성 추가 _ 구매건수

    - 구매건수 = 취급액/판매단가
    - 노이즈 데이터 처리 
    : 구매건수가 0보다 1보다 작으면서 판매단가가 50000인 경우는 제외(FAQ에 나옴)

```python
#구매건수 변수 추가
performance['cnt'] = performance['total_price']/performance['price']

print("구매건수 노이즈 처리 전 : ",len(performance))

#노이즈 데이터 처리
performance = performance[performance['cnt']>=1]
print("구매건수 노이즈 처리 후 : ",len(performance))

#print(len(performance[performance['total_price']==50000]))
```

### [특성 처리_2] 특성 추가 _ 할인여부

    - 구매건수 = 취급액/판매단가
    - 소수인 경우 -> "할인 받았다!"
    - (FAQ 내용)  고객이 상품을 구매할때 판매가를 그대로 지불하지 않고 할인쿠폰 적용, ARS할인, 일시불할인, 카드사할인 등 여러가지 경로로 할인된 금액을 지불하고있습니다.   (단, 결제시 적립금, 상품권 등을 사용하여 실결제금액이 바뀌는 경우는 해당사항 없음)

```python
def isDiscount(num):
    isdscnt = 1
    after_dot = int(str(num).split(".")[1])
    if after_dot ==0 : isdscnt=0
    else : pass
    return isdscnt 

#performance = performance.drop(['is_discount'], axis=1) 
performance['is_discount'] = performance['cnt'].apply(lambda x: isDiscount(x))

display(performance.head())
```

```python
display(performance[performance['is_discount']==0])
print("할인 받은 row:",len(performance[performance['is_discount']==1]))
print("할인 안받은 row:",len(performance[performance['is_discount']==0]))
# 약 5%가 할인을 안받고 거의 다 받았다

nondiscnt = performance[performance['is_discount']==0]
nondiscnt.prd_type_encoding.value_counts().plot(kind = 'bar')

for i in range(len(encoder_prdnm.classes_)):
    print(i,":",encoder_prdnm.classes_[i])
```

### [특성 처리_2] 특성 추가 _ 평일, 주말, 공휴일
- 평일 : 0
- 주말 + 공휴일 : 1

```python

display(performance)
```

```python
# 0(평일), 1(주말, 공휴일)
#전체 {1: [1], 2: [4, 5, 6], 3: [1], 5: [5, 8, 12, 15], 6: [6], 8: [15], 9: [12, 13, 14, 15], 10: [3, 9], 12: [25]}
#주말 제외 {1: [1], 2 : [4, 5, 6], 3: [1], 5: [8, 15], 6:[6], 8: [15], 9:[12,13], 10:[3,9], 12:[25] }

#평일, 주말로 1차 넣기
#performance = performance.drop(['is_holiday'], axis=1) 
performance['is_holiday_Sun'] = performance['dt_weekday1'].apply(lambda x: 1 if x==6 else 0)
performance['is_holiday_Sat'] = performance['dt_weekday1'].apply(lambda x: 1 if x>=5 else 0)
performance['is_holiday_Fri'] = performance['dt_weekday1'].apply(lambda x: 1 if x>=4 else 0)

# 공휴일로 2차 넣기
holiday_list = ['0101', '0204','0205','0206', '0508','0515','0606','0815','0912','0913','0914','0915','1003','1009','1225']
performance.reset_index(inplace = True)
performance = performance.drop(['index'], axis=1) 

for i in range(len(performance)):
    if performance['dt_mmdd'][i] in holiday_list:
        performance['is_holiday_Sat'][i]=1
        performance['is_holiday_Sun'][i]=1
        performance['is_holiday_Fri'][i]=1     

display(performance.head())
display(performance.is_holiday_Fri.value_counts().plot(kind = 'bar'))
```
