```
import pandas as pd
import numpy as np
from datetime import datetime
import time
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from matplotlib import font_manager, rc
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
mpl.rcParams['axes.unicode_minus'] = False
import warnings
warnings.filterwarnings('ignore')
import platform
plt.rc('font', family='NanumBarunGothic') 

from sklearn.ensemble import RandomForestRegressor

from scipy.sparse.linalg import svds
import matplotlib.font_manager as fm
#fm._rebuild()
plt.rc('font', family='NanumBarunGothic') 
```

```
data = pd.read_csv('/gdrive/My Drive/opt_data.csv')
data.head()
```

```
data = data[['방송일시','cat3','취급액']]
data.head()
```

```
data['방송일시'] = data['방송일시'].apply(lambda x: x[11:16])
data.head()
```

```
ori_time = data['방송일시'].unique()
```

```
data = data.groupby(['방송일시','cat3']).sum()[['취급액']].reset_index()
```

```
le_c = LabelEncoder()
le_c.fit(data['cat3'])

le_t = LabelEncoder()
le_t.fit(data['방송일시'])
```

```
data['cat'] = le_c.transform(data['cat3'])
#data['origin_cat'] = le_c.inverse_transform(data['cat'])
del data['cat3']

data['time'] = le_t.transform(data['방송일시'])
#data['origin_time'] = le_t.inverse_transform(data['time'])
del data['방송일시']
data.head()
```

```
data = data.pivot(index = 'time', columns = 'cat', values = '취급액')
data.head()
```

```
data.fillna(0, inplace=True)
data.head()
```

```
matrix = data.values
time_mean = np.mean(matrix, axis=1)
matrix_time_mean = matrix - time_mean.reshape(-1,1)
```

```
matrix
```

```
pd.DataFrame(matrix_time_mean, columns = data.columns).head()
```

```
# TruncatedSVD

U, sigma, Vt = svds(matrix_time_mean, k = 8)
print(U.shape)
print(sigma.shape)
print(Vt.shape)
```

```
sigma = np.diag(sigma)
sigma
```

```
svd_matrix = np.dot(np.dot(U, sigma), Vt) + time_mean.reshape(-1,1)
```

```
df_svd = pd.DataFrame(svd_matrix, columns = data.columns)
df_svd
```

```
candi = pd.DataFrame()
candi['time'] = df_svd.index
```

```
first = []
second = []
third = []

for i in df_svd.index:
  first.append(df_svd.loc[i].sort_values(ascending = False).index[0])
  second.append(df_svd.loc[i].sort_values(ascending = False).index[1])
  third.append(df_svd.loc[i].sort_values(ascending = False).index[2])
```

```
candi['first'] = first
candi['second'] = second
candi['third'] = third
candi.head()
```

```
candi['time'] = le_t.inverse_transform(candi['time'])
candi['first'] = le_c.inverse_transform(candi['first'])
candi['second'] = le_c.inverse_transform(candi['second'])
candi['third'] = le_c.inverse_transform(candi['third'])
```
