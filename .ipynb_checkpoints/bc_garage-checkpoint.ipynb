```python
# Train/Test-set의 고유한 마더코드와 상품코드를 살펴봅시다.
train_mcode = set(train['마더코드'])
test_mcode  = set(test['마더코드'])


# Train/Test-set 고유한 마더코드의 개수를 구함.
print(f"Train-set에 있는 고유한 마더코드 개수 : {len(train_mcode)}")
print(f"Test-set에 있는 고유한 마더코드 개수 : {len(test_mcode)}")

# Train/Test-set 모두에 포함되어있는 상품코드를 구함.
print('='* 80)
common_mcode = train_mcode.intersection(test_mcode)
print(f"Train/Test-set에 공통으로 포함되어 있는 마더코드 개수 : {len(common_mcode)}")
print(f"Train/Test-set에 공통으로 포함되어 있는 마더코드 개수 : {sorted(common_mcode)}")

# Train-set에만 있는 마더코드를 구함.
print('='* 80)
train_only_mcode = train_mcode.difference(test_mcode)
print(f"Train-set에만 있는 마더코드는 총 {len(train_only_mcode)}개 입니다.")
print(f"Train-set에만 있는 마더코드는 : {sorted(train_only_mcode)}")

# Test-set에만 있는 마더코드를 구함.
print('='* 80)
test_only_mcode = test_mcode.difference(train_mcode)
print(f"Test-set에만 있는 상품코드는 총 {len(test_only_mcode)}개 입니다.")
print(f"Test-set에만 있는 상품코드는 : {sorted(test_only_mcode)}")
```

```python
# 1. 일시불/무이자/없음

train.loc[train['product'].str.contains('일시불') == True, "product_plan"] = '1'
train.loc[train['product'].str.contains('\(일') == True, "product_plan"] = '1'
train.loc[train['product'].str.contains('무이자') == True, "product_plan"] = '2'
train.loc[train['product'].str.contains('\(무') == True, "product_plan"] = '2'
train['product_plan'] = train['product_plan'].fillna('0')
```

```python
# 2. 추가구성/단품구성

train.loc[train['product'].str.contains('\+') == True, "product_add"] = '1'
train['product_add'] = train['product_add'].fillna('0')
```

```python
# 3. 기타/삼성/LG

train.loc[train['product'].str.contains('삼성') == True, "product_brand"] = '1'
train.loc[train['product'].str.contains('LG') == True, "product_brand"] = '2'
train['product_brand'] = train['product_brand'].fillna('0')
```

```python
# 4. 세트구성/단품구성

train.loc[train['product'].str.contains('세트') == True, "product_set"] = '1'
train['product_set'] = train['product_set'].fillna('0')
```

```python
# 5. 여성/남성/없음

train.loc[train['product'].str.contains('여성') == True, "product_sex"] = '1'
train.loc[train['product'].str.contains('남성') == True, "product_sex"] = '2'
train['product_sex'] = train['product_sex'].fillna('0')
```

```python
# 일단 nan값을 노출칼럼에만 남기기 위해서 나머지 nan값 전처리 미리 해버리겠습니다.
train[' 취급액 '] = train[' 취급액 '].fillna('0')
train = train[train.상품군 != '무형']
```

찾았다. 먼저 난 값 없는 애들을 채우고, 그걸 다시 컨캣으로 합치고 인덱스 소팅 해준다음에, 난값 패딩해주면 끝.

인덱스가 꼬여서, 인덱스를 따로 놔두고, 리셋인덱스를 시켜서 포문 돌린 다음에, 인덱스를 다시 넣어주고, 합치겠습니다.

```python
# 인덱스 리셋하고 나중에 불러와주기 위해서 따로 저장해줍니다.
train_not_nan_index = train_not_nan.index
```

```python
# 인덱스 꼬여서 포문 안 돌아가기 때문에 리셋 인덱스 해줍니다.
train_not_nan_reset_index = train_not_nan.reset_index(drop=True)
```

채우는 원리는 다음과 같습니다. <br>
상품코드 순으로 앞에 것과 비교하는 것인데,<br>
1. 첫번째 상품코드는 자기의 노출(분) 값을 가져옵니다.
2. 두번째부터는 앞에것과 비교하는데, 앞에것과 상품코드가 같으면, 그 전 로우의 노출에 자기 노출을 더합니다. 그리고 카운트도 더합니다.
3. 앞에 것과 같지 않으면, 다시 자기의 노출을 가져오고 카운트는 초기화 시킵니다.

```python
cast_time = []
cast_count = []
time = 0.
count = 0.

for i in range(len(train_not_nan_reset_index)):
    if i == 0: # 첫번째 행은 이걸로 하십쇼.
        time = train_not_nan_reset_index['노출(분)'][i] # 더하지 않고, 자기 노출을 그대로 가져옵니다.
        count = 0 # count도 0으로 초기화 합니다.
        cast_time.append(time) # 그리고 어펜드.
        cast_count.append(count) # 그리고 어펜드.
    elif train_not_nan_reset_index['상품코드'][i] == train_not_nan_reset_index['상품코드'][i-1]: # 자기의 상품코드가 앞행의 상품코드와 같을 때
        time += train_not_nan_reset_index['노출(분)'][i] # 기존 누적된 노출에 자기 노출을 더한다.
        count += 1 # count도 1을 더해줍니다.
        cast_time.append(time) # 그리고 어펜드.
        cast_count.append(count) # 그리고 어펜드.
    else: # 근데 자기 상품코드와 앞행의 상품코드가 다르다면, 새로운 방송이 시작된 거겠죠?
        time = train_not_nan_reset_index['노출(분)'][i] # 더하지 않고, 자기 노출을 그대로 가져옵니다.
        count = 0 # count도 0으로 초기화 합니다.
        cast_time.append(time) # 그리고 어펜드.
        cast_count.append(count) # 그리고 어펜드.
```

```python
# 컬럼으로 박아줍니다.
train_not_nan_reset_index['cast_time'] = cast_time
train_not_nan_reset_index['cast_count'] = cast_count
train_not_nan_reset_index.head()
```

```python
# 원래 인덱스 다시 넣어줍니다.
train_not_nan_reset_index.index = train_not_nan_index
```

```python
# 컨캣으로 밑으로 붙여주고,
train_concat = pd.concat([train_not_nan_reset_index, train_yes_nan], axis=0)
```

```python
# 소트 인덱스 먼저 하고,
train_concat_sort_index = train_concat.sort_index()
```

```python
# ffill 로 채워줍니다. 끝. 이거 너무 힘들었다 진짜 레알.
train = train_concat_sort_index.fillna(method='ffill')
```

#### 동일 상품 이종 상품 역 가중치

미끼상품, 즉 동일한 방송에서 단가가 다른 제품을 판매할 때를 구분해주는 칼럼을 만들 겁니다.

원리는 다음과 같습니다.<br>
일단 데이터를 살펴보면, 법칙이(다행히도) 하나가 있습니다.<br>
바로 비싼 단가 순으로 내림차순이 되어 있다는 것입니다. 이를 활용하면

1. 가장 첫 1행은 k를 넣어줍니다. k=1 입니다.
2. 동일한 마더코드인데, n행의 단가보다 n-1행의 단가가 더 높다면 k+1 을 넣어줍니다.
3. 동일한 마더코드인데, n행의 단가보다 n-1행의 단가가 같다면 k를 넣어줍니다.
4. n-1행과 마더코드가 다르다면, k를 넣어줍니다.

이렇게 되면, 맨 처음, 단가 제일 높은 놈은 1이 들어가고,<br>
만약 그 다음 놈이 단가가 더 낮다면 2가 들어가고,<br>
또 낮으면 3이 들어가고 이런 식입니다.<br>
반대로 같으면 다시 1이 들어가고, 마더코드가 달라도 1이 들어가고.

이렇게 되면 무슨일이 일어나냐면
- 가장 비싼 제품:1
- 그 다음 비싼 제품:2
- 그 다음:3
- 새로운 제품이 나오면:1

여기서 저는 두 작업을 또 할 겁니다.

1. 주어진 값을 제곱한다. 그럼 1, 4, 9, 16 이런식으로 됩니다.
2. 이 값에 음수를 취해준다. 그럼 -1, -4, -9, -16 이렇게 될 겁니다.
3. -1은 0으로 바꿔준다. 그럼 0, -4, -9, -16이 될 겁니다.
4. 즉 0이 기준점이고, 저렴한 제품일수록 -(n**2)만큼 역가중치가 붙는겁니다.

```python
def stock_crawler(code):

    for c, i in zip(code, range(len(code))):
        
        url = 'https://finance.naver.com/item/sise_day.nhn' # URL은 고정
        result = []
        tdtext_list = ['날짜', c]

        for num in range(16, 43): # 페이지 범위 지정
            params = {'code' : c, # 지수 코드 지정
                      'page' : f'{num}'}

            resp = requests.get(url, params = params)
            soup = BeautifulSoup(resp.content, 'html.parser')

            for tr in soup.find_all('tr')[2:14]:
                new_dict = { }
                td_list = tr.find_all('td')
                if not td_list[0].text.strip():
                    continue
                for i in range(2):
                    new_dict[tdtext_list[i]] = td_list[i].text.strip()
                result.append(new_dict)

        globals()['stock_{}'.format(c)] = pd.DataFrame(result, columns = tdtext_list)
```

#### 동일 방송에서 다른 구성 상품을 판매하는지 여부 만들기

오차율이 매우 높은 퓨마 속옷은 3가지 상품을 함께 판매합니다. 사람들의 눈은 거의 비슷해서 하나의 상품이 다른 상품보다 100배는 팔립니다.<br>
문제는 이게 가격이 다 똑같아서 fake_weight 와는 조금 다릅니다. fake_weight는 동일 방송 다른 가격으로 수치를 뽑는데,<br>
푸마 속옷은 가격이 다 같아요. 이걸 1과 0으로 구분을 하려고 합니다.

비교해야 할 것은 다음과 같습니다.
1. 마더코드가 같아야 한다. 즉 동일 방송이어야 한다.
2. 판매단가가 같아야 한다. 즉 동일 가격이어야 한다.
3. 다만 상품코드는 달라야 한다. 즉 다른 구성의 상품이어야 한다.

어쩔 수 없습니다. 아래를 비교하고, 위를 비교해야 합니다. 맨 첫줄과 맨 아랫줄은 패스 합니다.

```python
def make_fake_weight2(train):
    fake_weight2 = []

    for i in range(len(train)):
        if i == 0: # 첫번째 행은 이걸로 하십쇼.
            fake_weight2.append('1') # 첫번째 행은 조건에 해당하니 강제 할당
            
        elif i == len(train)-1: # 마지막 행은 이걸로 하십쇼.
            fake_weight2.append('0') # 마지막 상품은 조건에 해당하지 않으니 강제 할당
            
        elif train['마더코드'][i] == train['마더코드'][i+1]: # 자기의 마더코드가 뒷행의 것과 같을 때
            if train['판매단가'][i] == train['판매단가'][i+1]: # 자기의 단가가 뒷행의 것과 같을 때(같으면 안됨)
                if train['상품코드'][i] != train['상품코드'][i+1]: # 마더코드와 단가는 맞는데, 상품코드가 다를 때
                    fake_weight2.append('1') # 1 넣기
                    
                else: # 상품코드까지 같을 때, 그건 그냥 동일 상품입니다.
                    fake_weight2.append('0') # 0 넣기
                    
            else: # 마더코드는 같은데 자기의 단가가 앞행의 것과 다르다면 이건 fake_weight1 입니다.
                fake_weight2.append('0') # 0 넣기
                
        elif train['마더코드'][i] == train['마더코드'][i-1]: # 자기의 마더코드가 앞행의 것과 같을 때
            if train['판매단가'][i] == train['판매단가'][i-1]: # 자기의 단가가 앞행의 것과 같을 때(같으면 안됨)
                if train['상품코드'][i] != train['상품코드'][i-1]: # 마더코드와 단가는 맞는데, 상품코드가 다를 때
                    fake_weight2.append('1') # 1 넣기
                    
                else: # 상품코드까지 같을 때, 그건 그냥 동일 상품입니다.
                    fake_weight2.append('0') # 0 넣기
                    
            else: # 마더코드는 같은데 자기의 단가가 앞행의 것과 다르다면 이건 fake_weight1 입니다.
                fake_weight2.append('0') # 0 넣기   
                
        else: # 앞뒤행과 모두 마더코드가 겹치지 않는다면.
            fake_weight2.append('0') # 0입니다.
            
    train['fake_weight2'] = fake_weight2
    
    return train
```
