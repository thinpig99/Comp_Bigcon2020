```python
# 기본 전처리 함수

def preprocessing_1(df):
    # 노출(분) 은 전 방송일시와 겹치면 Nan값으로 되어 있습니다.
    df['노출(분)'] = df['노출(분)'].fillna(method='ffill') # 바로 전 row의 값을 그냥 복사합니다.

    # 취급액은 주문이 없으면 0원으로 되어 있습니다.
    df[' 취급액 '] = df[' 취급액 '].fillna('0')
    
    # 상품군이 무형인 애들은 그냥 드랍시킵니다.
    df = df[df.상품군 != '무형']
    
    # 취급액과 판매단가 수치형 변수로 바꾸기.
    df[' 취급액 '] = df[' 취급액 '].apply(lambda x: x.replace(',', ''))
    df['판매단가'] = df['판매단가'].apply(lambda x: x.replace(',', ''))
    df[' 취급액 '] = df[' 취급액 '].astype(np.int32)
    df['판매단가'] = df['판매단가'].astype(np.int32)
    
    df = df.rename(columns = {'방송일시':'schedule',
                               '노출(분)':'term',
                               '마더코드':'main_code',
                               '상품코드':'sub_code',
                               '상품명':'product',
                               '상품군':'category',
                               '판매단가':'cost',
                               ' 취급액 ':'amount'})
    
    return df
```

```python
# 상품명 전처리 함수
def preprocessing_3(train):
    # 1. 일시불/무이자/없음

    train.loc[train['product'].str.contains('일시불') == True, "product_plan"] = '1'
    train.loc[train['product'].str.contains('\(일') == True, "product_plan"] = '1'
    train.loc[train['product'].str.contains('무이자') == True, "product_plan"] = '2'
    train.loc[train['product'].str.contains('\(무') == True, "product_plan"] = '2'
    train['product_plan'] = train['product_plan'].fillna('0')

    # 2. 추가구성/단품구성

    train.loc[train['product'].str.contains('\+') == True, "product_add"] = '1'
    train['product_add'] = train['product_add'].fillna('0')

    # 3. 기타/삼성/LG

    train.loc[train['product'].str.contains('삼성') == True, "product_brand"] = '1'
    train.loc[train['product'].str.contains('LG') == True, "product_brand"] = '2'
    train['product_brand'] = train['product_brand'].fillna('0')

    # 4. 세트구성/단품구성

    train.loc[train['product'].str.contains('세트') == True, "product_set"] = '1'
    train['product_set'] = train['product_set'].fillna('0')

    # 5. 여성/남성/없음

    train.loc[train['product'].str.contains('여성') == True, "product_sex"] = '1'
    train.loc[train['product'].str.contains('남성') == True, "product_sex"] = '2'
    train['product_sex'] = train['product_sex'].fillna('0')

    # category로 형변환해서 다이어트 시켜줍니다.

    train['product_plan'] = train['product_plan'].astype('category')
    train['product_add'] = train['product_add'].astype('category')
    train['product_brand'] = train['product_brand'].astype('category')
    train['product_set'] = train['product_set'].astype('category')
    train['product_sex'] = train['product_sex'].astype('category')
    
    return train
```

```python
train = preprocessing_1(train)
train = preprocessing_3(train)
train.head()
```

```python
# schedule 전처리 함수
def preprocessing_4(performance):
    
    performance['schedule'] = pd.to_datetime(performance['schedule'], errors='coerce')

    performance['schedule_yymmdd'] = performance['schedule'].apply(lambda x: str(x)[:10])
    performance['schedule_hhmmss'] = performance['schedule'].apply(lambda x: str(x)[11:])
    performance['seconds'] = performance['schedule_hhmmss'].apply(lambda x: int(x[:2])*60 + int(x[3:5]))
    
    return performance
```

```python
train = preprocessing_4(train)
train.head()
```

```python
ratings = tv.set_index('시간대')
ratings.mean(axis=1)
```

```python
ratings.drop('월화수목금토일02:00-01:59', axis=0, inplace=True)
```

```python
mean = ratings.mean(axis=1).values
ratings['mean'] = mean
ratings = ratings.reset_index()
ratings = ratings[['시간대','mean']]
ratings['시간대'] = ratings['시간대'].astype('datetime64[ns]')
ratings.set_index('시간대', inplace=True)
ratings.head()
```

```python
tmp = ratings[['mean']].resample('5T').mean().reset_index()
ratings = ratings.reset_index()
ratings = pd.merge(ratings, tmp, on='시간대', how='left')
ratings.head()
```

```python
ratings.drop('mean_x', axis = 1, inplace= True)
ratings.rename({'mean_y':'ratings_mean'}, axis=1, inplace=True)
ratings['시간대'] = ratings['시간대'].astype(str)
ratings.head()
```

```python
ratings['시간대'] = ratings['시간대'].apply(lambda x: x[-8:-3])
train['schedule_hhmm'] = train['schedule_hhmmss'].apply(lambda x: x[:-3])
performance = pd.merge(train, ratings, left_on = 'schedule_hhmm', right_on = '시간대', how='left')
del performance['시간대']
```

```python
performance.head()
```

```python
#performance = performance.fillna(method='pad')
performance.drop(['main_code','sub_code'], axis=1, inplace=True)
```

```python
performance = performance.sample(frac=1, random_state=529)
```

```python
train, test = train_test_split(performance, test_size=0.2, random_state=529)
print(train.shape)
print(test.shape)
```

```python
encoder_1 = LabelEncoder()
encoder_2 = LabelEncoder()

encoder_1.fit(train['product'].values)
encoder_2.fit(train['category'].values)

# encoded_1 = encoder_1.transform(train['상품명'].values)
# encoded_2 = encoder_2.transform(train['상품군'].values)

train['product_encoding'] = encoder_1.transform(train['product'])
train['category_encoding'] = encoder_2.transform(train['category'])

for label in np.unique(test['product']):
    if label not in encoder_1.classes_: # unseen label 데이터인 경우( )
        encoder_1.classes_ = np.append(encoder_1.classes_, label) # 미처리 시 ValueError발생

for label in np.unique(test['category']):
    if label not in encoder_2.classes_: # unseen label 데이터인 경우( )
        encoder_2.classes_ = np.append(encoder_2.classes_, label) # 미처리 시 ValueError발생


test['product_encoding'] = encoder_1.transform(test['product'])
test['category_encoding'] = encoder_2.transform(test['category'])

display(train.head(), test.head())
```

```python
train_x = train[['term','cost','seconds','product_encoding','category_encoding','ratings_mean']].values
train_y = train['amount'].values
test_x = test[['term','cost','seconds','product_encoding','category_encoding','ratings_mean']].values
test_y = test['amount'].values
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```
